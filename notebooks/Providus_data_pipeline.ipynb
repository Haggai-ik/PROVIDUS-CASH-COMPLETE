```python
def clean_pipeline(df):
    df.FORCAST_DATE=pd.to_datetime(df.FORCAST_DATE)
    Day_name=df.FORCAST_DATE.dt.day_name()
    df['Day_name']=Day_name
    weekend=['Thursday', 'Friday', 'Monday', 'Wednesday', 'Tuesday']
    df=df[df.Day_name.isin(weekend)]
    return df
df=clean_pipeline(df)
```

```python
DEPOSIT_PREDICTION_TABLE= pd.DataFrame({'REFERENCE_DATE': df.REFERENCE_DATE,
                           'FORECAST_DATE': df.FORCAST_DATE,
                           'BRANCH_CODE':df.Branch_code,
                           'TRAN_CRNCY_CODE':df.tran_crncy_code,
                           'TARGET_CREDIT':df.Target_Credit,
                           'DEP_FORECAST':c_predictions,
                           'DAY_MONTH':df.DAY_MONTH,
                           'MONTH':df.Month,
                           'PREPUBHOL':df.PREBUBHOL,
                           'POSTPUBHOL':df.POSTBUBHOL})
WITHDRAWAL_PREDICTION_TABLE= pd.DataFrame({'REFERENCE_DATE': df1.REFERENCE_DATE,
                           'FORECAST_DATE': df1.FORCAST_DATE,
                           'BRANCH_CODE':df1.Branch_code,
                           'TRAN_CRNCY_CODE':df1.tran_crncy_code,
                           'TARGET_DEBIT':df1.Target_Debit,
                           'WITH_FORECAST':d_predictions,
                           'DAY_MONTH':df1.DAY_MONTH,
                           'MONTH':df1.Month,
                           'PREPUBHOL':df1.PREBUBHOL,
                           'POSTPUBHOL':df1.POSTBUBHOL})
```

```python
DEPOSIT_PREDICTION_TABLE['DEP_FORECAST']=np.where((DEPOSIT_PREDICTION_TABLE['TARGET_CREDIT'] ==0) & (DEPOSIT_PREDICTION_TABLE['DEP_FORECAST'] < 0) , 0, DEPOSIT_PREDICTION_TABLE['DEP_FORECAST'])
WITHDRAWAL_PREDICTION_TABLE['WITH_FORECAST']=np.where((WITHDRAWAL_PREDICTION_TABLE['TARGET_DEBIT'] ==0) & (WITHDRAWAL_PREDICTION_TABLE['WITH_FORECAST'] < 0) , 0, WITHDRAWAL_PREDICTION_TABLE['WITH_FORECAST'])
```

```python
q = """select b.REFERENCE_DATE,
       b.FORECAST_DATE,
       b.Branch_code,
       a.TRAN_CRNCY_CODE,
       a.Target_Credit,
       b.Target_Debit,
       c._AVAIL_BAL VAULT_BALANCE,
       a.DEP_FORECAST DEPOSIT_PREDICTION,
       b.WITH_FORECAST WITHDRAWAL_PREDICTION,
       (c._AVAIL_BAL +a.DEP_FORECAST-b.WITH_FORECAST) PREDICT_VAUILT_BAL,
       b.DAY_MONTH,b.MONTH MONTH,b.PREPUBHOL
       from DEPOSIT_PREDICTION_TABLE a
       inner join WITHDRAWAL_PREDICTION_TABLE b on a.REFERENCE_DATE= b.REFERENCE_DATE
       inner join vault c on a.REFERENCE_DATE=c.BANK_DATE
       where a.Branch_code =b.Branch_code and a.TRAN_CRNCY_CODE=b.TRAN_CRNCY_CODE and a.Branch_code=c.BRA_CODE and a.TRAN_CRNCY_CODE=c.CUR_CODE
       ;"""
Final_prediction_table = pysqldf(q)
```
